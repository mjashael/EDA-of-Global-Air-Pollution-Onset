## Univariate Analysis

```python
# Assuming you have a DataFrame 'df' with columns 'AQI Value' and 'CO AQI Value'
col = ['AQI Value', 'CO AQI Value']

# Create a figure with subplots
fig = plt.figure(figsize=(15, 5))

# Iterate through each column and create a histogram plot
for i in range(len(col)):
    plt.subplot(1, 5, i + 1)  # 1 row, 5 columns
    plt.title(col[i])  # Set the title for each subplot
    sns.histplot(df, x=col[i])  # Create the histogram

plt.tight_layout()  # Adjust subplot spacing
plt.show()  # Display the plot
```

## Bivariate Analysis

```python
# Assuming you have a DataFrame 'df' with columns 'AQI Value', 'CO AQI Value',
# 'Ozone AQI Value', 'NO2 AQI Value', and 'PM2.5 AQI Value'
col = ['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

# Create a new figure with a specified size (15 inches wide and 10 inches tall)
fig = plt.figure(figsize=(15, 10))

# Iterate through each column in the 'col' list
for i in range(len(col)):
    # Create a subplot grid with 2 rows and 3 columns
    plt.subplot(2, 3, i + 1)  # The third argument specifies the position of the current subplot
    plt.title(col[i])  # Set the title for the current subplot
    # Create a box plot using Seaborn
    # - 'data=df': Use the DataFrame 'df' as the data source
    # - 'x=df[col[i]]': Specify the data for the x-axis (values from the current column)
    # - 'y=df['AQI Category']': Specify the data for the y-axis (values from the 'AQI Category' column)
    sns.boxplot(data=df, x=df[col[i]], y=df['AQI Category'])

# Adjust the layout of subplots to prevent overlapping
plt.tight_layout()

# Display the entire figure containing all subplots
plt.show()
```

## Multivariate Analysis
