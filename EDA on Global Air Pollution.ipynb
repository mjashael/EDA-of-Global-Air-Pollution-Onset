## Univariate Analysis

```python
# Assuming you have a DataFrame 'df' with columns 'AQI Value' and 'CO AQI Value'
col = ['AQI Value', 'CO AQI Value']

# Create a figure with subplots
fig = plt.figure(figsize=(15, 5))

# Iterate through each column and create a histogram plot
for i in range(len(col)):
    plt.subplot(1, 5, i + 1)  # 1 row, 5 columns
    plt.title(col[i])  # Set the title for each subplot
    sns.histplot(df, x=col[i])  # Create the histogram

plt.tight_layout()  # Adjust subplot spacing
plt.show()  # Display the plot
```

```python
# Assuming you have a DataFrame 'df' with columns 'Ozone AQI Value', 'NO2 AQI Value', and 'PM2.5 AQI Value'
col = ['Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

# Create a figure with subplots
fig = plt.figure(figsize=(15, 5))

# Iterate through each column and create a histogram plot
for i in range(len(col)):
    plt.subplot(1, 5, i + 1)  # 1 row, 5 columns
    plt.title(col[i])  # Set the title for each subplot
    sns.histplot(df, x=col[i])  # Create the histogram

plt.tight_layout()  # Adjust subplot spacing
plt.show()  # Display the plot
```

## Bivariate Analysis

```python
fig = plt.figure(figsize=(15,10))

for i in range(len(col)):
    plt.subplot(2,3,i+1)
    plt.title(col[i])
    sns.boxplot(data=df,x=df[col[i]],y=df['AQI Category'])
plt.tight_layout()
plt.show()
```

## Multivariate Analysis

```python
# Set the color palette
sns.set_palette("Pastel2")

# Assuming 'df' is your DataFrame
plt.figure(figsize=(12, 10))

# Using Seaborn to create a pair plot with the specified color palette
sns.pairplot(df)

plt.suptitle('Pair Plot for DataFrame')
plt.tight_layout()
plt.show()

```

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(15, 10))
 
# Using Seaborn to create a heatmap
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)
 
plt.title('Correlation Heatmap')
plt.show()
```
